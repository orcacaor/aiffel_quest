```python

#예외처리도 하면 되는데 안했음



#제품:제품번호 딕셔너리
po_num = {
    "사과 포션" : 1,
    "포도 포션" : 2,
    "수박 포션" : 3,
    "참외 포션" : 4
}



po_num_rev = {}
for (i, k) in po_num.items() :
    po_num_rev[k] = i



#제품:가격 딕셔너리
po_price = {
    "사과 포션" : 100,
    "포도 포션" : 200,
    "수박 포션" : 300,
    "참외 포션" : 400
}


#제품:재고 딕셔너리
po_stock = {
    "사과 포션" : 40,
    "포도 포션" : 30,
    "수박 포션" : 20,
    "참외 포션" : 10
}



bill_sum = 0 #누적 주문 총액 변수 선언



#메뉴판 무한루프 지점
while True :
    
    #누적 주문 총액 검사
    if bill_sum > 10000 :
        po_price = {
            "사과 포션" : int(100 * 0.9),
            "포도 포션" : int(200 * 0.9),
            "수박 포션" : int(300 * 0.9),
            "참외 포션" : int(400 * 0.9)
        }
        print("누적 주문 총액 10,000원 돌파! 10% 할인")

    #메뉴판 출력
    print("-" * 25)
    print("번호   이름   가격  재고")

    for i in po_num :

        print(" " + str(po_num[i]) + "  " + i + " " + str(po_price[i]) + " " + str(po_stock[i]))

    print("-" * 25)


    #입장 질문
    enter = input("주문하시겠습니까? 맞으면 '네', 나가려면 'ㅂㅂ' 입력")

    if enter == "네" :

        
        order = {} #번호:재고. 주문서 딕셔너리 선언
    
        #주문 무한루프 지점
        while True :

            bill = 0 #주문 총액 변수 선언
            
            
            num = int(input("주문할 제품 번호를 입력하라. 주문을 완료하려면 '0'을 입력하라"))

            #주문 완료 경우
            if num == 0 :
                
                print("주문 내용은")
                #주문완료시 재고 반영 및 주문 총액 계산
                for (i, k) in order.items() :
                    
                    
                    po_stock[po_num_rev[i]] -= order[i]
                    #print(po_stock[po_num_rev[i]])
                    
                    bill += po_price[po_num_rev[i]] * order[i]
                    bill_sum += bill
                    
                    print(po_num_rev[i] + " " + str(order[i]) + "개" + " " + str(po_price[po_num_rev[i]] * order[i]) + "원"  )
                    
                    
                
                print()
                print("주문 총액: " + str(bill))
                print("누적 주문 총액: " + str(bill_sum))
                    
                break
            
            else : #재고 입력

                stock = int(input("갯수를 입력하세요"))
                order[num] = stock 



    else :
        print("다시는 오지마세요")
        break


```
