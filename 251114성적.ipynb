```python

#학생 등록 여부 조회 기능
#신규 학생 성적 입력 기능
#학생별 성적 따로 보관 기능
#최저, 최고, 평균 점수 출력 기능
#반복해서 조회 및 입력 기능
#종료 기능


#더 완벽하게 하려면 0~100점 범위 초과 입력시 예외처리도 해야하지만 잡다한 것은 넘어감


subject = ["국어", "영어", "수학", "사회", "과학"] #과목 리스트

student = {} #학생목록 딕셔너리. 이름(str):성적묶음(dic)



while True : #무한루프
    

    stu_name = input("이름을 입력하세요. 종료하려면 exit를 입력하세요.") #이름 입력

    
    
    if stu_name in student : #학생 목록에 존재하는지 확인
        
        score = student[stu_name] #해당하는 성적묶음 딕셔너리 호출

        for (i, k) in score.items() : #과목별 점수 출력
            print(i + str(k))
        
        min = 100 #최저 점수 출력용 변수 선언
        min_sub = "최저 과목"
        max = 0 #최고 점수 출력용 변수 선언
        max_sub = "최고 과목"

        sum = 0 #점수 총합 계산용 변수 선언
        sub_num = 0 #과목 숫자 카운트 변수 선언

        for (i, k) in score.items() :
            sum += k
            sub_num += 1

            if k <= min :
                min = k
                min_sub = i

            if k >= max :
                max = k
                max_sub = i

        average = sum / sub_num #평균 점수 출력용 변수 생성
        print("최저 점수: " + min_sub + str(min)) #최저 점수 출력
        print("최고 점수: " + max_sub + str(max)) #최고 점수 출력
        print("평균 점수: " + str(average)) #평균 점수 출력
        



    elif stu_name == "exit" : #무한루프 종료 트리거
        print("성적 프로그램 종료")
        break



    else : #존재하지 않으면 성적 입력받기
        score = {} #성적묶음 딕셔너리. 과목(str):성적(int) #성적묶음 딕셔너리 생성.
        student[stu_name] = score #학생목록 딕셔너리에 해당 이름 등록

        for i in subject : #과목별 점수 입력
            score[i] = int(input(i + "성적을 입력하라"))

        for (i, k) in score.items() : #과목별 점수 출력
            print(i + str(k))


        min = 100 #최저 점수 출력용 변수 선언
        min_sub = "최저 과목"
        max = 0 #최고 점수 출력용 변수 선언
        max_sub = "최고 과목"


        sum = 0 #점수 총합 계산용 변수 선언
        sub_num = 0 #과목 숫자 카운트 변수 선언

        for (i, k) in score.items() :
            sum += k
            sub_num += 1

            if k <= min :
                min = k
                min_sub = i

            if k >= max :
                max = k
                max_sub = i

        average = sum / sub_num #평균 점수 출력용 변수 생성
        print("최저 점수: " + min_sub + str(min)) #최저 점수 출력
        print("최고 점수: " + max_sub + str(max)) #최고 점수 출력
        print("평균 점수: " + str(average)) #평균 점수 출력

    
    
  

    
    
    
    
```
