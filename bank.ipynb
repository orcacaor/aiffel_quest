```python
import random

#요구사항
#1. Account 클래스
#이름: 모두은행, 예금주, 계좌번호, 잔액
#계좌번호 3-2-6자리 랜덤. 첫자리 0 금지
#클래스 변수로 인스턴스 갯수 기록

class Account:
    bank = "모두은행"
    acc_num = 0 #인스턴스 갯수

    def __init__(self, name):
        self.name = name
        self.bal = 0
        self.deposit_count = 0

        #계좌번호 생성
        num_list = ""
        num_list += str(random.randint(1,9))  #첫자리 0 제외 난수
        for n in range(0, 10): #나머지 10자리 난수
            num_list += str(random.randint(0,9))

        # print(num_list) #11자리 확인
        num_new = num_list[:3] + "-" + num_list[3:5] + "-" + num_list[5:] #하이픈
        # print(num_new) #결과 확인


        self.num = num_new #생성 함수
        self.record_history = [] #리스트

        Account.acc_num += 1



    #히스토리 기록용 메서드
    def record(self, event, amount, bal):
        self.record_history.append([event, amount, bal])



#2. 입금 메서드
#입금은 1원 이상
#입금 5회마다 잔고 기준 1% 이자
#입금 내역 기록 (입금만 따로 출력 가능?)


    def deposit(self, amount):
        if amount < 1:
          print("입금은 1원 이상만 가능합니다. 다시 시도해주세요.")
          return

        self.deposit_count += 1
        self.bal += amount

        self.record("입금", amount, int(self.bal))

        if self.deposit_count % 5 == 0:
          self.bal += self.bal * 0.01
          print(f"모두은행을 {self.deposit_count}회 이용해주셔서 감사합니다. 이자 1%({self.bal * 0.01})가 적립되었습니다.")
          self.record("이자지급", amount, int(self.bal))
        print(f"{amount:,}원 입금하였습니다. 잔액은 {self.bal:,}원 입니다.")


#3. 출금 메서드
#한계: 계좌 잔고
#출금 내역 기록 (출금만 따로??)

    def withdraw(self, amount):
        if amount > self.bal:
          print("잔액이 부족합니다.")
          return

        self.bal -= amount
        self.record("출금", amount, int(self.bal))

#4. get_account_num() 메서드
#인스턴스 갯수 출력
#인스턴스가 만들어질 때마다 숫자 올려줘야함

    def get_account_num(self):
        return f"생성된 계좌 수: {Account.acc_num}개" #확인 필요

#5. history() 메서드
#입금 내역과 출금 내역 출력
#일단 시간순으로 구현하자
#입금따로 출금따로 -- 옵션

    #리스트 리스트로 구현하자
    def history(self):
        # print("히스토리 호출 성공")
        # print(self.record_history)
        for idx, (event, amount, bal) in enumerate(self.record_history):
            # print("포문 들어옴")
            print(f"{idx + 1}회: {event:<6}\t금액: {amount:<6}\t잔액: {bal}")

    def get_num(self):
        return f"계좌번호는 {self.num}입니다."



acc1 = Account("김이름1")
acc1.deposit(1000)
acc1.deposit(2000)
acc1.withdraw(2000)
acc1.deposit(2000)
acc1.deposit(2000)
acc1.deposit(2000)
acc1.deposit(2000000)
acc1.withdraw(3000)

acc1.history()

acc1.get_num()

acc2 = Account("김이름2")
acc3 = Account("김이름3")
acc4 = Account("김이름4")

acc4.get_account_num()
```
